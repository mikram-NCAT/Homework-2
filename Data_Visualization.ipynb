A histogram is a type of graph that shows how data is distributed. It works by dividing your data into intervals (called bins) and counting how many data points fall into each bin. Then it displays these counts as bars

##Advanced Subplots with Shared Colorbar

```python
# === Styled 22 heatmaps with shared colorbar, unified scaling, and cleaner grid ===

# Figure + grid (tighter spacing and room for colorbar)
fig = plt.figure(figsize=(13, 9), facecolor="#f9f9f9")
gs = gridspec.GridSpec(
    2, 2,
    width_ratios=[1, 1],
    height_ratios=[1, 1],
    wspace=0.25, hspace=0.30
)

# Data
data1 = np.random.uniform(low=0, high=1, size=(10, 10))
data2 = np.random.uniform(low=0, high=1, size=(10, 10))
data3 = np.random.uniform(low=0, high=1, size=(10, 10))
data4 = np.random.uniform(low=0, high=1, size=(10, 10))

# Shared color scale across all heatmaps (so colors mean the same thing)
vmin = min(d.min() for d in [data1, data2, data3, data4])
vmax = max(d.max() for d in [data1, data2, data3, data4])

# Font styles
title_font = {"fontname": "DejaVu Sans Mono", "fontsize": 15, "fontweight": "bold", "color": "#222"}
label_font = {"fontname": "DejaVu Sans Mono", "fontsize": 12, "color": "#333"}
tick_color = "#444"

# Axes
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

# A nice divergent colormap (good for lowhigh contrast)
cmap = "RdYlBu_r"

# Helper to style each heatmap axis
def style_heatmap(ax, data, title):
    im = ax.imshow(
        data, cmap=cmap, vmin=vmin, vmax=vmax,
        interpolation="nearest", aspect="equal"
    )
    ax.set_title(title, **title_font, pad=8)
    ax.set_xlabel("X-axis", **label_font, labelpad=6)
    ax.set_ylabel("Y-axis", **label_font, labelpad=6)

    # Light gridlines between cells
    ax.set_xticks(np.arange(-0.5, data.shape[1], 1), minor=True)
    ax.set_yticks(np.arange(-0.5, data.shape[0], 1), minor=True)
    ax.grid(which="minor", color="white", linewidth=0.8)
    ax.tick_params(colors=tick_color, labelsize=10)
    # Show only major ticks (0..n-1)
    ax.set_xticks(np.arange(0, data.shape[1], 2))
    ax.set_yticks(np.arange(0, data.shape[0], 2))
    # Matrix-style top-left origin
    ax.invert_yaxis()
    return im

im1 = style_heatmap(ax1, data1, "Heatmap 1")
im2 = style_heatmap(ax2, data2, "Heatmap 2")
im3 = style_heatmap(ax3, data3, "Heatmap 3")
im4 = style_heatmap(ax4, data4, "Heatmap 4")

# Shared colorbar
cbar_ax = fig.add_axes([0.90, 0.12, 0.02, 0.76])  # [left, bottom, width, height]
cbar = fig.colorbar(im1, cax=cbar_ax)
cbar.set_label("Color Intensity", **label_font)
cbar.ax.tick_params(colors=tick_color, labelsize=10)

# Layout (leave space for colorbar)
plt.tight_layout(rect=[0, 0, 0.88, 1])
plt.show()
```

##3D Plotting

```python
# === 3D Surface Plot with Style Enhancements ===

# Generate 3D data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create figure and 3D axis
fig = plt.figure(figsize=(12, 9), facecolor='#f9f9f9')
ax = fig.add_subplot(111, projection='3d')

# Plot 3D surface with refined style
surf = ax.plot_surface(
    x, y, z,
    cmap='coolwarm',          # Vibrant and intuitive color palette
    edgecolor='none',         # Removes mesh lines for a smooth surface
    antialiased=True,
    alpha=0.9,
    linewidth=0
)

# Add contour projections on walls
ax.contourf(x, y, z, zdir='z', offset=-1.2, cmap='coolwarm', alpha=0.6)

# Customize title and labels
title_font = {'fontname': 'DejaVu Sans Mono', 'fontsize': 16, 'fontweight': 'bold', 'color': '#222'}
label_font = {'fontname': 'DejaVu Sans Mono', 'fontsize': 12, 'color': '#333'}

ax.set_title('3D Surface Plot of sin((x + y))', **title_font, pad=15)
ax.set_xlabel('X-axis', **label_font, labelpad=10)
ax.set_ylabel('Y-axis', **label_font, labelpad=10)
ax.set_zlabel('Z-axis', **label_font, labelpad=10)

# Adjust 3D viewing angle
ax.view_init(elev=30, azim=135)  # elevation, azimuth

# Add colorbar
cbar = fig.colorbar(surf, shrink=0.6, aspect=15, pad=0.1)
cbar.set_label('Z Value (sin((x + y)))', **label_font)

# Layout and display
plt.tight_layout()
plt.show()
```
